# 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

- 당뇨병 데이터 셋 

```python
from sklearn.datasets import load_diabetes
diabetes= load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()
```

- 당뇨병 (target) vs BMI

```python
X = df.bmi.values.reshape(-1,1) #시리즈를 넘파이의 어레이로 만들기 위해 values를 사용 그다음 reshape을 해줌 
y = df.target
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
weight, bias = lr.coef_, lr.intercept_
rss = lr._residues
weight[0],bias, rss
```

```python
lr.score(X,y)
```

- train/test data 분리
- test_size = 0.1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state= 2021
)
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
pred = lr.predict(X_test)
pred[:4]
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```

```python
from sklearn.metrics import mean_squared_error #1/n 시그마 ( )2제곱
mse = mean_squared_error(y_test,pred)
rmse = np.sqrt(mse)
mse, rmse
```

- 그래프로 확인

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.show()
```

```python
w, b =  lr.coef_, lr.intercept_
plt.scatter(X_train, y_train)
plt.plot([-0.1,0.18],[-0.1*w+b, 0.18*w+b],'r','lw=3')
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.1,0.18],[-0.1*w+b, 0.18*w+b],'r','lw=3')
plt.grid()
plt.show()
```

- 당뇨병 vs BP

```python
X = df.bp.values.reshape(-1,1) #시리즈를 넘파이의 어레이로 만들기 위해 values를 사용 그다음 reshape을 해줌 
y = df.target
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state= 2021
)
```

```python
diabetes.data[:,3:4].shape # 행도 열도 슬라이싱 하면 2차원이 됨 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data[:,3:4], diabetes.target, test_size=0.1, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
w, b = lr.coef_[0], lr.intercept_
rss = lr._residues
w, b, rss
```

```python
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)
mse, rmse, r2
```

```python
plt.scatter(X_train, y_train)
plt.plot([-0.12,0.13], [-0.12*w+b, 0.13*w+b], 'r', lw=3)
plt.title('Diabetes vs BP')
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.12,0.13], [-0.12*w+b, 0.13*w+b], 'r', lw=3)
plt.title('Diabetes vs BP')
plt.grid()
plt.show()
```

### 회귀 표본 만들어 연습하기

```python
from sklearn.datasets import make_regression
X, y = make_regression(
    n_samples=100, n_features=1, n_targets=1, random_state=2021
)
```

```python
plt.scatter(X, y)
plt.grid()
plt.show()
```

```python
X,y = make_regression(n_samples=100, n_features=1,n_targets=1, noise=100, random_state=2021)
plt.scatter(X, y)
plt.grid()
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X, y)
w, b = lr.coef_[0], lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot([-3,4], [-3*w+b, 4*w+b], 'r', lw=3)
plt.grid()
plt.show()
```

```python
r2 = lr.score(X, y)
w, b, r2
```
