#  군집화(Clustering) - K-Means Algorithm

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
scaler = StandardScaler()
iris_std = scaler.fit_transform(iris.data)

pca = PCA(n_components=2)
iris_pca = pca.fit_transform(iris_std)
```

```python
iris_df = pd.DataFrame(iris.data, columns=['sl','sw','pl','pw'])
iris_df['pca_x'] = iris_pca[:, 0]
iris_df['pca_y'] = iris_pca[:, 1]
iris_df.head()
```

### K-Means 모델 생성/학습

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=2021)
```

```python
kmeans.fit(iris_std)
kmeans.labels_
```

```python
iris_df['target'] = iris.target
iris_df['cluster'] = kmeans.labels_
iris_df.tail()
```

```python
clustering_result = iris_df.groupby(['target', 'cluster']).count()
clustering_result
```

### 데이터 시각화

- 군집화한 결과를 차원축소하여 시각화

```python
markers = ['^','s','o']

for i, marker in enumerate(markers):
    x_data = iris_df[iris_df['cluster']==i]['pca_x']
    y_data = iris_df[iris_df['cluster']==i]['pca_y']
    plt.scatter(x_data, y_data, marker=marker)

plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

- 원 데이터를 차원축소하여 시각화

```python
for i, marker in enumerate(markers):
    x_data = iris_df[iris_df['target']==i]['pca_x']
    y_data = iris_df[iris_df['target']==i]['pca_y']
    plt.scatter(x_data, y_data, marker=marker, label=iris.target_names[i])

plt.legend()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

- 비교그림 

```python
fig, asx = plt.subplots(figsize=(12,6), ncols=2)
columns = ['target', 'cluster']

for k, column in enumerate(columns):
    ax = asx[k]
    for i, marker in enumerate(markers):
        x_data = iris_df[iris_df[column]==i]['pca_x']
        y_data = iris_df[iris_df[column]==i]['pca_y']
        ax.scatter(x_data, y_data, marker=marker, label=iris.target_names[i])

    ax.set_xlabel('PCA 1')
    if k == 0:  # target인 경우 
        ax.set_title('Original Data')
        ax.set_ylabel('PCA 2')
        ax.legend()
    else:       # cluster인 경우 
        ax.set_title('Clustered Data')
```
